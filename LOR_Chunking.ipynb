```python
import nltk
import docx2txt
from nltk.chunk import *
from nltk.chunk.util import *
from nltk.chunk.regexp import *
```

```python
lor_sample_1_text = docx2txt.process("Dataset\\LoR Sample 1.docx")
lor_sample_2_text = docx2txt.process("Dataset\\LoR Sample 2.docx")  
```

```python
print(lor_sample_1_text)
```

```python
print(lor_sample_2_text)
```

```python
lor_1_tokens = nltk.word_tokenize(lor_sample_1_text)
lor_2_tokens = nltk.word_tokenize(lor_sample_2_text)
```

```python
print(f"No of tokens in LoR 1: {len(lor_1_tokens)}")
print(f"No of tokens in LoR 2: {len(lor_2_tokens)}")
```

```python
pos_tags_lor_1 = nltk.pos_tag(lor_1_tokens)
pos_tags_lor_2 = nltk.pos_tag(lor_2_tokens)
```

```python
print(pos_tags_lor_1)
```

```python
chunk_rule = ChunkRule("<PRP>+<RB>*(<VBD>|<VBP>)*<VBZ>*<VBN>*(<PRP>|<PRP$>)*<RB>*<JJ>?", "Chunk Verb Phrases")
chink_rule = ChinkRule("<CD|CC|IN|TO|\,|\.>", "Chink on TO/prepositions")

chunk_parser = RegexpChunkParser([chunk_rule, chink_rule],chunk_label='VP')
chunked_lor_1 = chunk_parser.parse(pos_tags_lor_1)
```

```python
print(chunked_lor_1)
```

```python
chunked_lor_1.draw()
```

```python
ner_lor_1 = nltk.ne_chunk(pos_tags_lor_1)
print(ner_lor_1)
```

```python
ner_lor_1.draw()
```

```python
chunk_rule = ChunkRule("<DT>?<JJ>*<NN>?", "Chunk Noun Phrases")

chink_rule = ChinkRule("<CC|IN|TO|\.>", "Chink on TO/prepositions")

chunk_parser = RegexpChunkParser([chunk_rule, chink_rule],chunk_label='NP')
chunked_lor_2 = chunk_parser.parse(pos_tags_lor_2)
```

```python
print(chunked_lor_2)
```

```python
chunked_lor_2.draw()
```

```python
ner_lor_2 = nltk.ne_chunk(pos_tags_lor_2)
print(ner_lor_2)
```

```python
ner_lor_2.draw()
```
