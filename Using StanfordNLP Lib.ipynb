```python
# stanza.download('hi')
import stanza
```

```python
with open ("Dataset\\Dyn_book_1_m_Hindi.txt" , 'r',encoding='utf16', errors='strict') as f:
    hindi_text = f.read()

print(hindi_text)
```

```python
nlp = stanza.Pipeline(lang='hi', processors='tokenize, pos, lemma, depparse')
```

```python
document = nlp(hindi_text)
```

```python
hindi_tokens = []
for i, sentence in enumerate(document.sentences):
    sentence_tokens = []
    [sentence_tokens.append(token.text) for token in sentence.tokens]
    hindi_tokens.append(sentence_tokens)
```

```python
print(f"No of Tokens : {len(hindi_tokens)}\n")
for sent_tok in hindi_tokens:
    print(f'{sent_tok}')
```

```python
hindi_pos_tags = [(word.text, word.upos, word.xpos) for sent in document.sentences for word in sent.words]
print(hindi_pos_tags)
```

```python
hindi_words_lemma = [(word.text, word.lemma) for sent in document.sentences for word in sent.words]
print(hindi_words_lemma)
```

```python
hindi_words_feats = [(word.text, word.feats) for sent in document.sentences for word in sent.words]
print(hindi_words_feats)
```
